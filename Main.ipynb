```python
grow_week = pd.DataFrame({'Year': [2020, 2021, 2022, 2023],
                   'grow_rub': [11965.66, 1844.33, 2693.12, 797.08],
                         'grow_%': [86.0, 18.0, 43.0, 33.0]})
grow_week
```

```python
grow_day = pd.DataFrame({'Year': [2020, 2021, 2022, 2023],
                   'grow_rub': [84824.85, 13115.73, 18819.25, 5383.77],
                         'grow_%': [88.0, 19.0, 44.0, 32.0]})
grow_day
```

```python
plt.rcParams["figure.figsize"] = (12, 8)

# Создание фигуры и осей для каждого графика
fig, axes = plt.subplots(nrows=2, ncols=2)

# Отрисовка первого графика на верхнем левом углу
axes[0, 0].plot(grow_week['Year'], grow_week['grow_rub'])
axes[0, 0].set_title('Прибыль в рублях, инв. раз в неделю')
axes[0, 0].grid()
axes[0, 0].scatter(grow_week['Year'], grow_week['grow_rub'])
for i in range(len(grow_week['Year'])):
    axes[0 ,0].text(grow_week.at[i,'Year'], grow_week.at[i,'grow_rub']+150, 
                    f"{grow_week.at[i,'grow_rub']}", fontsize=11) # Подписываем значения точек

# Отрисовка второго графика на верхнем правом углу
axes[0, 1].plot(grow_week['Year'], grow_week['grow_%'])
axes[0, 1].set_title('Прибыль в процентах, инв. раз в неделю')
axes[0, 1].grid()
axes[0, 1].scatter(grow_week['Year'], grow_week['grow_%'])
for i in range(len(grow_week['Year'])):
    axes[0 ,1].text(grow_week.at[i,'Year'], grow_week.at[i,'grow_%']+2, 
                    f"{grow_week.at[i,'grow_%']}", fontsize=11) # Подписываем значения точек

# Отрисовка третьего графика на нижнем левом углу
axes[1, 0].plot(grow_day['Year'], grow_day['grow_rub'])
axes[1 ,0].set_title('Прибыль в рублях, инв. раз в день')
axes[1, 0].grid()
axes[1, 0].scatter(grow_day['Year'], grow_day['grow_rub'])
for i in range(len(grow_week['Year'])):
    axes[1 ,0].text(grow_day.at[i,'Year'], grow_day.at[i,'grow_rub']+150, 
                    f"{grow_day.at[i,'grow_rub']}", fontsize=11) # Подписываем значения точек

# Отрисовка четвертого графика на нижнем правом углу
axes[1 ,1].plot(grow_day['Year'], grow_day['grow_%'])
axes[1 ,1].set_title('Прибыль в процентах, инв. раз в день')
axes[1, 1].grid()
axes[1, 1].scatter(grow_day['Year'], grow_day['grow_%'])
for i in range(len(grow_week['Year'])):
    axes[1 ,1].text(grow_day.at[i,'Year'], grow_day.at[i,'grow_%']+2, 
                    f"{grow_day.at[i,'grow_%']}", fontsize=11) # Подписываем значения точек

# Установка компактного расположения подписей между подграфиками
plt.tight_layout()

# Показать все графики
plt.show()
```
