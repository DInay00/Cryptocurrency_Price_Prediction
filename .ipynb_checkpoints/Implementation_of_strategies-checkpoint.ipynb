```python
def Permanent_purchase_week(start_time, end_time):
    """
    FFFF
    """
    
    api_key = 'mWoLPGkpNCPw5LlDU8plBgxHQ62QIm9eQk8VvtRS260pFX04HHj8lghP0SrJFoJd'
    api_secret = 'CTZAWA6Bklwibnz9bYqAtAXivkLVtu4aoD3bXbM3BG0soZL2ZRUGsKuWirOxJsBu'

    symbol1 = "BTCUSDT"
    symbol2 = "USDTRUB"

    client_btc = Client(api_key, api_secret)
    client_usdt = Client(api_key, api_secret)

    klines_btc = client_btc.get_historical_klines(symbol1, Client.KLINE_INTERVAL_1WEEK, start_time, end_time)
    klines_usdt=client_usdt.get_historical_klines(symbol2, Client.KLINE_INTERVAL_1WEEK, start_time, end_time)


    df_btc = pd.DataFrame(klines_btc, columns=['Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close Time', 'Quote Asset Volume',
                                   'Number of Trades', 'Taker Buy Base Asset Volume', 'Taker Buy Quote Asset Volume', 'Ignore'])

    df_usdt = pd.DataFrame(klines_usdt, columns=['Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close Time', 'Quote Asset Volume',
                                   'Number of Trades', 'Taker Buy Base Asset Volume', 'Taker Buy Quote Asset Volume', 'Ignore'])

    df_btc = df_btc.astype(float)
    df_usdt = df_usdt.astype(float)
    
    df_btc['date'] = (df_btc['Time']/1000).astype('datetime64[s]')
    
    df_btc.reset_index(inplace=True, drop=True)
    df_usdt.reset_index(inplace=True, drop=True)

    df_btc.drop(columns=['Time','Close Time','Ignore', 
                 'Taker Buy Quote Asset Volume', 'Taker Buy Base Asset Volume',
                'Volume', 'Quote Asset Volume', 'Number of Trades',
                    'Open', 'Low', 'Close'], inplace=True)

    df_usdt.drop(columns=['Time','Close Time','Ignore', 
                 'Taker Buy Quote Asset Volume', 'Taker Buy Base Asset Volume',
                'Volume', 'Quote Asset Volume', 'Number of Trades',
                     'Open', 'Low', 'Close'], inplace=True)

    df_btc.rename(columns={'High': 'High_btc'}, inplace=True)
    df_usdt.rename(columns={'High': 'High_usdt'}, inplace=True)

    df_strategy1 = pd.concat([df_btc, df_usdt], axis=1).copy()

    df_strategy1['bitcoins_received'] = 1/df_strategy1['High_btc']

    df_strategy1['Sum of bitcoins'] = 0.0
    for i in range(1, len(df_strategy1)):
        # Получение значений предыдущих строк из обоих стоблцов
        bitcoins_received = df_strategy1.at[i-1,'bitcoins_received']
        Number_of_bitcoins = df_strategy1.at[i-1,'Sum of bitcoins']
    
        x = bitcoins_received + Number_of_bitcoins
        df_strategy1.at[i,'Sum of bitcoins'] = x

    
    df_strategy1['spent_rub'] = 0.0
    for i in range(1, len(df_strategy1)):
        # Получение значений предыдущих строк из обоих стоблцов
        High_usdt = df_strategy1.at[i-1,'High_usdt']
        spent_rub = df_strategy1.at[i-1,'spent_rub']
    
        x = High_usdt + spent_rub
        df_strategy1.at[i,'spent_rub'] = x

    df_strategy1['received_rub'] = df_strategy1['Sum of bitcoins'] * df_strategy1['High_btc'] * df_strategy1['High_usdt']
    profit = df_strategy1.at[len(df_strategy1)-1,'received_rub'] - df_strategy1.at[len(df_strategy1)-1, 'spent_rub']
    profit_percent = -1 + df_strategy1.at[len(df_strategy1)-1,'received_rub'] / df_strategy1.at[len(df_strategy1)-1, 'spent_rub']
    
    return df_strategy1, profit, profit_percent
```
